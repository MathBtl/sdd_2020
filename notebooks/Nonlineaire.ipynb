## Wisdom of crowds

```python
N = 3
points = np.random.uniform(size=(5, N))

#my_lss = ['--', ':', '-.']
#plt.rcParams['axes.prop_cycle'] = ("cycler('ls', %s)" % my_lss)

fig = plt.figure(figsize=(8, 8))

for ix in range(points.shape[0]):
    escalier = np.sort(points[ix, :])
    xvals = [0, 0]
    xxx = np.repeat(escalier, 2)
    xvals.extend(xxx)
    xvals.extend([1])
    
    escalier += 0.1 * (np.random.random(size=N)-0.5)
    yvals = [0]
    yyy = np.repeat(escalier, 2)
    yvals.extend(yyy)
    yvals.extend([1, 1])

    plt.plot(xvals, yvals, ':', color='k', lw=2) 
plt.plot([-0.05, 0.85], [0.15, 1.05], lw=3, ls='-')
t=plt.axis('off')

plt.savefig('../poly/figures/nonlin/crowd_diagonal.pdf', bbox_inches='tight')
```
